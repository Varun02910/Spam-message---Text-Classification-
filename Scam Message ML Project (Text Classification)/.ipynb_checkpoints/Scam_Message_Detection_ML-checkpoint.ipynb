### Data Preprocessing

```python
#Converting the Scam or ham into a numerical value
# taking 1 = scam && 0 = ham (not a scam)

#Modifying Data.. (modified_data)
#We Can use mapping also for the best practice.

data["Category"].loc[data["Category"] == "spam"] = 1
data["Category"].loc[data["Category"] == "ham"] = 0

```

##### Splitting the data into test and train set

```python
#Feature Extraction and spliting of the dataset into test and train 

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer 

#Spliting of data into train and test set
X_train,X_test, y_train, y_test = train_test_split(data["Message"], data["Category"], test_size=0.20, random_state=36)
```

```python
#text Vectorization

vectorizer = TfidfVectorizer(stop_words='english')

X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
X_train_tfidf, X_test_tfidf
```

```python
#Text vectorization in the form of dataframe
tfidf_df = pd.DataFrame(X_test_tfidf.toarray(), columns = vectorizer.get_feature_names_out())

tfidf_df
```

### Model training using SVM (SVC) for text Classification

```python
y_train = y_train.astype(int)
y_test = y_test.astype(int)
```

```python
#importing ml estimator and fitting the train data into it
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

svc_model = SVC()

svc_model.fit(X_train_tfidf, y_train)

```

```python
y_pred = svc_model.predict(X_test_tfidf)

print( classification_report(y_test, y_pred))
print ("Confusion Matrix: \n", confusion_matrix(y_test, y_pred))
```
